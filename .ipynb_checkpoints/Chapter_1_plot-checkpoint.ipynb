```python
from matplotlib import pyplot as plt
```

```python
years = (2010,2011,2012,2013,2014,2015,2016,2017,2018,2019)
salaries = (63000,64000,65000,68000,70000,75000,76000,79000,80000,82000)
prices = (40000,42000,43000,45000,48000,49000,52000,53000,54000,55000)


plt.plot(years , salaries, 'o--y' , label = 'Salaries')
plt.plot(years , prices , label = 'Prices',color='g',linestyle='-.',marker='D')
plt.title('Salaries vs. Prices')
plt.xlabel('Years')
plt.ylabel('USD')
plt.legend()
plt.grid()
```
